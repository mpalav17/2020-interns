```python
import pandas as pd
import numpy as np
file=pd.read_json('data.json')
file.shape
dates=[]
for i in range(0,255):
    dates.append(file.index.date[i])
```

```python
import datetime
tr=0
st=0
et=0
print('enter start date')
sdy=input('year')
sdm=input('month')
sdd=input('day')
print('enter end date')
edy=input('year')
edm=input('month')
edd=input('day')
print('enter no. of currencies to be compared with EUR')
nc=input()
curr=[]
for i in range(0,int(nc)):
    curr.append(input('enter curreny in abbreviation'))
for d in dates:
    tr+=1
    if d<datetime.date(int(sdy),int(sdm),int(sdd)):
        st=tr
    if d<datetime.date(int(edy),int(edm),int(edd)):
        et=tr
import matplotlib.pyplot as plt
df=pd.DataFrame(file)
date_list=[]
for j in range(st,et):
    date_list.append(df.index[j].date())
plt.figure(figsize=(20,10))
for types in curr:
    types=types.upper()
    curr_list=[]
    for k in range(st,et):
        curr_list.append(df['rates'][k][types])
    lrate=pd.read_json('latest-rates.json')
    add=str(lrate['rates'][types])
    typess=types+' '+add
    plt.plot(date_list,curr_list,label=typess)
plt.title('Graph of Currencies against Datewise rates')    
plt.xlabel('Dates')
plt.ylabel('Rate')
plt.legend()
plt.show()
```
